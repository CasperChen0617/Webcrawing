```python
import requests
```

```python
site = "https://query1.finance.yahoo.com/v8/finance/chart/2330.TW?period1=0&period2=1549258857&interval=1d&events=history&=hP2rOschxO0"
response = requests.get(site)
response.text

# 瀏覽器通常是用「get」的方法，
# 而此網址必須要用post方法，才能正確的得到資料。
```

```python
# with open('file.csv', 'w') as f:
#     f.writelines(response.text)
import json
import pandas as pd
import numpy as np
data = json.loads(response.text)

# 用dataframe取股價各值
df = pd.DataFrame(data['chart']['result'][0]['indicators']['quote'][0], index=pd.to_datetime(data['chart']['result'][0]['timestamp'],unit='s'), columns=['open','high','low','close','volume'])
# 日期索引值欄位名稱設定
df=df.rename_axis("Date")
# 開高低收設定
df=df.rename_axis("stock_price", axis="columns")

# 發現此文字檔非常大…，這種格式叫做 json，
# 我們先把文字檔轉成 python 的 dictionary 跟 list 的結構
# 並且將資料轉成dataframe
# df = pd.read_csv('file.csv', index_col='date', parse_dates=['Dates'])
df.head()
```
